# CAP. 1 Enfrentando o dia a dia 1: VIDA SOCIAL E MÚSICA 

1. **Como devemos agir diante do sistema do mundo, segundo essas passagens de Efésios 4:21-32?**

Deixar práticas e comportamentos corruptos, adotar uma mentalidade alinhada aos ensinamentos de Cristo, ser honesto e transparente com todos, não permitir que a ira me leve ao pecado e resolver conflitos rapidamente, me comunicar de maneira positiva e construtiva. Remover amargura, cólera, ira, gritaria, blasfêmia e malícia. Trabalhar de forma honesta e ajudar os necessitados.

2. **Depois de aceitarmos a Cristo, precisamos aprender a conduzir os nossos relacionamentos. Como devemos responder às perguntas dos nossos familiares, amigos e etc?**

Com paciência, respeito e honestidade, respondendo aos questionamentos da nova jornada que eu estou começando. Usar não só a Bíblia, mas também testemunhos que vivi ou de pessoas que conheci. Com humildade, tendo empatia e compreensão de que nem todos vão aceitar minha nova vida.

3. **Qual é o propósito de Satanás por meio da música?**

Desviar a adoração de Deus.

4. **A música foi criada para dar glória e honra a Deus, mas o mal a tem usado para desviar o cristão do caminho da salvação. Como podemos lidar com a má influência da música?**

Escolher músicas edificantes, desenvolver discernimento: avaliar as letras e mensagens das músicas à luz dos ensinamentos bíblicos, limitar a exposição ou reduzir a exposição a músicas com mensagens negativas ou contrárias à fé cristã.

5. **Quais são os frutos da música gospel em nossas vidas?**

Edificação Espiritual: Fortalece a fé e promove o crescimento espiritual, ajudando a internalizar os ensinamentos bíblicos.  
Adoração e Louvor: Facilita a adoração a Deus, proporcionando uma forma de expressar gratidão e reverência.  
Conforto e Esperança: Oferece consolo em momentos difíceis e renova a esperança através de mensagens de amor e promessa de Deus.  
Inspira mudanças positivas no comportamento e na atitude, alinhando a vida com os princípios cristãos.  
Comunhão e Unidade: Promove a união entre os cristãos, criando um senso de comunidade e fraternidade.

6. **Complete:**

"Antes, santificai em vossos corações a Cristo como Senhor, e estai sempre preparados para responder com mansidão e temor a todo aquele que vos pedir a razão da esperança em vós." (1 Pedro 3:15)

7. **O que acontece com as pessoas que passam o dia escutando música secular (música do "mundo")?**

A música tem um impacto significativo nas emoções e pode alterar o humor e o comportamento. Músicas com mensagens depressivas, agressivas ou promotoras de rebeldia podem afetar negativamente o bem-estar emocional e mental. Uma exposição constante a conteúdos seculares pode tornar uma pessoa menos sensível às orientações do Espírito Santo, tornando mais difícil discernir a vontade de Deus em suas vidas.

# CAP. 2 Enfrentando o dia a dia II: Vencendo as Tentações

1. **O que é o "mundo"?**<p>
O Universo Criado: Refere-se ao mundo físico, a criação de Deus, incluindo a terra, os céus, e tudo o que existe neles.
        "No princípio criou Deus os céus e a terra." (Gênesis 1:1)<p>
A Humanidade: Pode referir-se às pessoas que habitam a terra.
        "Porque Deus amou o mundo de tal maneira que deu o seu Filho unigênito, para que todo aquele que nele crê não pereça, mas tenha a vida eterna." (João 3:16)<p>
O Sistema de Valores Oposto a Deus: Refere-se ao sistema de valores, pensamentos e comportamentos que são contrários à vontade de Deus.
        "Não ameis o mundo, nem o que no mundo há. Se alguém ama o mundo, o amor do Pai não está nele." (1 João 2:15)<p>
O Império de Satanás: O "mundo" também pode simbolizar o domínio de Satanás, que é considerado o "príncipe deste mundo".
        "Agora é o juízo deste mundo; agora será expulso o príncipe deste mundo." (João 12:31)

2. **Como o cristão é afetado pelo mundo?**<p>
Tentação e Pecado: Enfrenta tentações que podem afastá-lo de Deus (1 João 2:16).<p>
Perseguição: Pode sofrer oposição e perseguição por sua fé (João 15:18; Mateus 5:10).<p>
Pressão Social: Há pressão para se conformar aos padrões do mundo, em vez de seguir os ensinamentos de Cristo (Romanos 12:2).<p>
Testemunho: É chamado a ser luz e sal no mundo, vivendo de maneira que glorifique a Deus (Mateus 5:14; Marcos 16:15).<p>
Conflito Espiritual: Está envolvido em uma batalha espiritual contra as forças do mal (Efésios 6:12).<p>

3. **Por que é importante ter atitudes diferentes para vencer o mundo e suas tentações?**<p>
Santificação: Ser separado e santo em ações e pensamentos (1 Pedro 1:15-16).<p>
Testemunho: Demonstrar a transformação de Cristo aos outros (Mateus 5:16).<p>
Resistência ao Pecado: Evitar cair em tentações que afastam de Deus (Tiago 4:7).<p>
Transformação da Mente: Renovar a mente para discernir a vontade de Deus (Romanos 12:2).<p>
Proteção Espiritual: Proteger-se contra influências negativas (Efésios 6:11).<p>

4. **Deus criou o homem, a mulher e o sexo. Qual foi o Seu propósito ao criar o sexo?**<p>
Procriação: Permitir a reprodução humana (Gênesis 1:28).<p>
União: Fortalecer o vínculo emocional e físico entre marido e mulher (Gênesis 2:24).<p>
Prazer: Proporcionar prazer e satisfação dentro do casamento (Provérbios 5:18-19).<p>
Proteção Contra a Imoralidade: Prover um contexto seguro e santo para expressar a sexualidade (1 Coríntios 7:2-3).<p>

  - Bíblia Católica: 46 livros no Antigo Testamento (inclui 7 livros adicionais).

Coleção oficial de livros considerados inspirados e autoritativos pela comunidade religiosa.

  - **Inspiração Divina**: Reconhecer a inspiração pelo Espírito Santo.
  - **Autoridade Apostólica**: Relacionamento com os apóstolos ou seus colaboradores.
  - **Consistência Doutrinária**: Alinhamento com a fé cristã e ensinamentos.

  - **Comunhão com Deus**: Profundidade no relacionamento com Deus.

# Cap.4 A arma do cristão II: Oração e Jejum

1. **O que significa orar?**

- Orar é comunicar-se com Deus, expressando louvor, gratidão, pedidos, confissões e intercessões.

2. **Como deve começar a nossa oração? (II Crônicas 7:14)**

- Começa com humildade, arrependimento e busca sincera a Deus: "Se o meu povo, que se chama pelo meu nome, se humilhar, e orar..."

3. **Cite quais são os passos para uma oração eficaz.**

 **Humildade**: Reconhecer a necessidade de Deus.
 **Fé**: Crer que Deus ouve e responde.
 **Confissão**: Arrepender-se dos pecados.
 **Persistência**: Continuar orando sem desanimar.
 **Gratidão**: Agradecer a Deus por Suas bênçãos.

4. **Quais devem ser os motivos das nossas orações?**

- Louvor e adoração, gratidão, intercessão pelos outros, petição por necessidades pessoais, confissão de pecados e orientação divina.

5. **Explique o que é jejum e quando devemos jejuar.**

- **Jejum**: Abstinência de alimentos por um período para focar na oração e buscar a Deus.
- **Quando Jejuar**: Em tempos de decisão, arrependimento, crise espiritual ou necessidade especial de orientação divina.

6. **Qual é a motivação correta para jejuarmos?**

- Buscar intimidade com Deus, purificação espiritual e alinhamento com a vontade de Deus, não para exibição ou ganho pessoal.

**7. Complete:**

# CAP. 5 As Características de Deus
1. **O que é um atributo?**
- Um atributo é uma característica ou qualidade inerente a uma pessoa ou coisa.

2. **Quais são os atributos exclusivos de Deus?**
- Onipotência, onisciência, onipresença, eternidade e imutabilidade.

- Deus é amor e nos ama incondicionalmente. Somos chamados a amar a Deus e ao próximo.

- Retidão é a qualidade de ser moralmente correto e justo em todas as ações e pensamentos.

- **Justiça**: Agir conforme a lei moral e a vontade de Deus.
- **Necessária**: Reflete o caráter de Deus e é fundamental para viver uma vida cristã que honra a Deus e serve ao próximo.

6. **Como Deus quer que sejamos?**
- Deus quer que sejamos santos, justos, amorosos, humildes e obedientes à Sua vontade.

# CAP. 07  O desenvolvimeto da Fé com a condução do Espírito Santo
1. **Ao entregar o coração a Jesus, a pessoa passa a viver um processo de mudança. O que significa santificação?**
Santificação, para um cristão, refere-se ao processo contínuo pelo qual uma pessoa é transformada para se tornar mais semelhante a Cristo. Envolve a purificação do pecado, a renovação da mente e a crescente conformação ao caráter de Deus. Este processo é marcado por uma busca constante pela santidade e pelo alinhamento com a vontade de Deus.

2. **⁠Quem é o Espírito Santo?**
O Espírito Santo é a terceira pessoa da Trindade cristã, coexistindo eternamente com Deus Pai e Deus Filho (Jesus Cristo). Ele é o Consolador prometido por Jesus aos seus seguidores, que habita nos crentes para guiar, ensinar, capacitar e santificar.

3. **⁠Complete:**
"Eleitos segundo a presciência de **Deus Pai**, em santificação do **Espírito**, para a obediência e aspersão **do sangue** de **Jesus Cristo**: Graça e paz vos sejam multiplicadas." (I Pedro 1:2)

4. **Como podemos receber o Espírito Santo?** 
Os cristãos acreditam que recebem o Espírito Santo através da fé em Jesus Cristo como seu Salvador pessoal. Isso acontece quando alguém entrega sua vida a Cristo, aceita o perdão dos pecados e busca uma vida de obediência e comunhão com Deus.

5. **O Espírito Santo age de diversas maneiras. Preocupado em nos tornar aptos para o reino de Deus, Ele transforma nosso caráter e nos concede poderes sobrenaturais para efetuarmos nosso serviço ao Senhor. A presença do Espírito Santo deixa frutos em nossas vidas. Com base em Gálatas 5:22, relacione a segunda coluna de acordo com a primeira.**

 1. Amor - ( ) É tanto emoção e sentimento, como uma decisão, uma opção de vida, que exige determinação e disciplina.
 2. Alegria - ( ) Ela não advém de uma situação externa e transitória, mas tem motivos permanentes e eternos.
 3. Paz - ( ) É a capacidade de estar contente em qualquer circunstância. Recusar-se a viver em murmurações e lamentações.
 4. Longanimidade - ( ) Constância nas afeições, sentimentos e perseverança nas alianças.
 5. Benignidade - ( ) É a tendência, a disposição, a intenção de fazer o bem. É conhecer as limitações e falhas do outro, mas mesmo assim agradar-se em satisfazê-lo.
 6. Bondade - ( ) Qualidade do caráter de quem é bom. Tolerância, compreensão. Sempre praticante de coisas boas.
 7. Fidelidade - ( ) É firmeza de ânimo, disposição para suportar crises e dificuldades, mantendo o equilíbrio e a fé, sem se entregar ao desespero e à depressão.
 8. Mansidão - ( ) É ser suave no trato com os outros. É entregar na mão de Deus seus direitos e sua defesa.
 9. Domínio Próprio - ( ) Poder resistir à vontade do pecado, tem disciplina e vontade. É o senhorio de si mesmo.

6. **De que maneira podemos ser cheios do Espírito Santo de Deus?**
   - **Oração e Pedido:** Uma forma fundamental é pedir a Deus em oração para ser cheio do Espírito Santo. Jesus ensinou que Deus dá o Espírito Santo àqueles que pedem (Lucas 11:13).
   - **Obediência à Palavra:** Viver de acordo com os ensinamentos bíblicos e obedecer aos mandamentos de Deus é essencial. A obediência demonstra um coração disposto e submisso à liderança do Espírito.
   - **Arrependimento e Confissão:** Confessar pecados e se arrepender deles é vital, pois o pecado pode impedir a plenitude do Espírito Santo. Um coração limpo e arrependido é mais receptivo à Sua presença.
   - **Louvor e Adoração:** Engajar-se em louvor e adoração genuínos convida a presença do Espírito Santo. O ambiente de adoração pode trazer uma maior consciência da presença e do poder do Espírito.
   - **Comunhão com outros Crentes:** Participar da comunhão com outros crentes, especialmente em contextos de culto e oração coletiva, pode fortalecer a experiência do Espírito Santo.
   - **Estudo da Bíblia:** Estudar as Escrituras regularmente ajuda a renovar a mente e a alinhar-se com a vontade de Deus, permitindo uma maior sensibilidade à direção do Espírito Santo.

7. **Como ocorre o Batismo com o Espírito Santo?**
   - **Oração e Imposição de Mãos:** Muitas vezes, o batismo com o Espírito Santo ocorre através da oração e, em alguns casos, da imposição de mãos por parte de outros crentes ou líderes espirituais. Em Atos 8:17, os apóstolos Pedro e João impuseram as mãos sobre os crentes e eles receberam o Espírito Santo.
   - **Fé e Expectativa:** Ter fé e expectativa de receber o Espírito Santo é crucial. Em Atos 2:38, Pedro instruiu os crentes a se arrependerem e serem batizados em nome de Jesus Cristo para receberem o dom do Espírito Santo.
   - **Manifestação de Dons Espirituais:** O batismo no Espírito Santo muitas vezes é acompanhado por manifestações de dons espirituais, como falar em línguas, profecia ou outras manifestações descritas em 1 Coríntios 12:7-11.
   - **Busca Sincera:** Um desejo sincero e fervoroso de receber o Espírito Santo é fundamental. Em Atos 1:14, vemos os discípulos em oração constante, buscando o cumprimento da promessa do Espírito Santo.
   - **Eventos Sobrenaturais:** O batismo no Espírito Santo pode ocorrer em momentos de adoração intensa, durante a pregação da Palavra, ou até mesmo em momentos de crise pessoal onde há um clamor sincero por Deus.
   - **Entrega Completa:** Uma rendição completa a Deus e uma disposição de permitir que o Espírito Santo opere livremente na vida pessoal é crucial. A disposição de seguir a liderança do Espírito e permitir Sua obra transformadora facilita essa experiência.

# CAP. 08 Fé

1. **Explique com as suas palavras o que é fé?**
   Fé é uma confiança profunda e inabalável em algo ou alguém, mesmo sem evidências físicas ou provas concretas. No contexto cristão, é acreditar e confiar em Deus e em Suas promessas, mesmo quando não podemos ver ou entender tudo. A fé nos motiva a viver de acordo com as verdades que acreditamos, sustentando-nos em tempos de dificuldade e guiando nossas ações diárias.

2. **Como ministério, qual é a nossa declaração de fé?**
   A declaração de fé de um ministério é um resumo das crenças fundamentais e doutrinas que guiam a comunidade de fé. Embora cada ministério possa ter sua própria declaração específica, geralmente inclui crenças sobre a Trindade (Deus Pai, Filho e Espírito Santo), a salvação através de Jesus Cristo, a autoridade da Bíblia, a importância da santificação e o papel da Igreja. 

3. **Complete:**
   "Ora, **a fé** é o firme fundamento das coisas que **se esperam**, e a prova das coisas que **não se veem**." Hebreus 11:1

4. **Não basta apenas pensar com fé e ser crédulo. É necessário também agir em fé e ter atitudes de fé. Explique o que seria ter atitudes de fé?**
   Ter atitudes de fé significa colocar em prática as crenças e confiança que se tem em Deus. Isso envolve tomar decisões e ações que refletem a confiança em Deus, mesmo em situações incertas ou difíceis. Exemplos de atitudes de fé incluem orar com confiança, servir aos outros, testemunhar sobre a fé, tomar decisões morais difíceis baseadas em princípios bíblicos e perseverar em meio a desafios, confiando que Deus está no controle.

5. **O que é preciso para descobrirmos a diferença que a fé pode fazer em nossas vidas?**
   Para descobrir a diferença que a fé pode fazer em nossas vidas, é necessário:
   - **Experimentar a fé pessoalmente:** Isso envolve buscar um relacionamento pessoal com Deus através de oração, leitura da Bíblia e participação em uma comunidade de fé.
   - **Praticar a fé diariamente:** Colocar em prática os ensinamentos bíblicos e confiar em Deus em todas as áreas da vida.
   - **Testemunhar as mudanças:** Observar e refletir sobre como a fé impacta positivamente as atitudes, decisões, relacionamentos e a paz interior.
   - **Perseverar em desafios:** Confiar em Deus mesmo em tempos difíceis, reconhecendo que Ele trabalha em todas as coisas para o bem daqueles que O amam.

6. **No que vale a pena você depositar a nossa fé?**
   Vale a pena depositar nossa fé em Deus, pois Ele é fiel, imutável e todo-poderoso. A fé deve estar centrada em Jesus Cristo e nas promessas de Deus reveladas na Bíblia. Também é importante confiar no plano de Deus para nossas vidas, na Sua provisão e proteção, e na esperança da vida eterna.

7. **Qual o resultado de uma vida de fé?**
   O resultado de uma vida de fé inclui:
   - **Paz interior:** A confiança em Deus traz uma paz que supera as circunstâncias.
   - **Força e Resiliência:** A fé proporciona força para enfrentar desafios e adversidades.
   - **Transformação de caráter:** A vida de fé resulta em um caráter moldado segundo os valores e virtudes cristãs.
   - **Comunhão com Deus:** A fé aproxima-nos de Deus, promovendo uma relação íntima e pessoal com Ele.
   - **Esperança:** A fé oferece uma esperança firme e segura no futuro, especialmente na promessa de vida eterna.
   - **Impacto positivo:** Uma vida de fé influencia positivamente os outros, através do testemunho pessoal e do serviço aos demais.
   - **Direção e Propósito:** A fé dá sentido e propósito à vida, guiando decisões e ações diárias de acordo com a vontade de Deus.


# CAP. 09 Obediência 

1. **Definição de Obediência**:
   Obediência é a ação de seguir ou cumprir uma ordem, instrução ou comando dado por uma autoridade. No contexto bíblico, refere-se ao ato de seguir as instruções e mandamentos de Deus, demonstrando fidelidade e submissão à Sua vontade.

2. **Tipos de Obediência**:
   - **Obediência Imediata**: Seguir as instruções de Deus sem hesitação, no momento em que são dadas.
   - **Obediência Parcial**: Cumprir apenas parte do que foi ordenado, sem seguir integralmente as instruções de Deus.
   - **Obediência Tardia**: Cumprir uma ordem, mas de maneira atrasada, o que pode ser visto como desobediência.
   - **Obediência Total**: Seguir as instruções de Deus completamente e de maneira imediata, sem questionamento ou hesitação.

3. **Importância de Evitar a Obediência Tardia**:
   Obediência tardia pode ser vista como desobediência, pois não cumpre os comandos de Deus no tempo determinado. Deus deseja que sejamos prontos e diligentes em obedecer Sua vontade. Obediência atrasada pode levar a consequências negativas e demonstra falta de confiança e respeito pela autoridade divina.

4. **Versículo Completo**:
   "Porém Samuel disse: Tem porventura o Senhor tanto prazer em holocaustos e sacrifícios como em que se obedeça à palavra do SENHOR? Eis que o obedecer é melhor do que o sacrificar; e o atender, melhor do que a gordura de carneiros." - 1 Samuel 15:22

5. **Exemplos de Obediência e Desobediência na Bíblia**:
   - **Obediência**: 
     - Abraão ao obedecer a Deus e estar disposto a sacrificar seu filho Isaque (Gênesis 22).
     - Noé ao construir a arca seguindo as instruções de Deus, salvando sua família do dilúvio (Gênesis 6-9).
   - **Desobediência**:
     - Adão e Eva ao comerem do fruto proibido, desobedecendo a Deus (Gênesis 3).
     - O rei Saul ao poupar Agague e o melhor do gado, contrariando a ordem de Deus de destruir tudo (1 Samuel 15).

6. **Consequências da Desobediência no Velho Testamento**:
   - Expulsão do Jardim do Éden para Adão e Eva (Gênesis 3).
   - Morte no dilúvio para aqueles que não seguiram as advertências de Deus por meio de Noé (Gênesis 7).
   - Destruição de Sodoma e Gomorra devido à desobediência e maldade dos habitantes (Gênesis 19).
   - Derrota e morte do rei Saul, que perdeu o trono por desobedecer a ordem de Deus (1 Samuel 15:23).

7. **Caminhos Dentro da Vontade de Deus**:
   - **Santificação**: Viver uma vida separada para Deus, buscando ser santo como Ele é santo.
   - **Amor e Serviço ao Próximo**: Demonstrar o amor de Deus ao servir e cuidar dos outros.
   - **Fé e Confiança**: Confiar plenamente em Deus e viver pela fé, não pelas circunstâncias.
   - **Obediência à Palavra**: Seguir os mandamentos e ensinamentos da Bíblia como guia para a vida.
   - **Submissão ao Espírito Santo**: Ser guiado pelo Espírito Santo, permitindo que Ele dirija cada passo da vida.

# CAP. 10 Bênção e Maldição
1. **Conceitos**:
   - **a. Bênção**: Bênção é um favor ou graça concedida por Deus, que traz prosperidade, paz, proteção, e bem-estar. No contexto bíblico, uma bênção é uma manifestação da bondade e do cuidado de Deus, resultando em um impacto positivo sobre a vida da pessoa ou comunidade.
   - **b. Maldição**: Maldição é uma consequência negativa, seja espiritual ou física, que ocorre como resultado da desobediência ou rebelião contra Deus. É a retirada da proteção divina, levando a dificuldades, enfermidades, ou outros tipos de aflições.

2. **O que traz as maldições?**
   - As maldições são frequentemente resultantes da desobediência às leis e mandamentos de Deus. A Bíblia ensina que afastar-se de Deus, praticar idolatria, imoralidade, injustiça, e rejeitar a Sua palavra pode trazer maldições. Outras causas incluem pactos com o mal, práticas ocultas, e ações que vão contra a vontade divina.

3. **Importância de Identificar as Maldições**:
   - Identificar as maldições é importante porque permite que a pessoa ou comunidade reconheça áreas da vida onde a influência negativa está atuando. Isso possibilita buscar soluções espirituais, como arrependimento e oração, para quebrar essas maldições e restaurar a bênção de Deus. Também ajuda a interromper ciclos de sofrimento que podem estar sendo perpetuados.

4. **Passos para Quebrar as Maldições**:
   - **Arrependimento**: Confessar pecados e desobediências que podem ter dado origem às maldições.
   - **Renúncia**: Rejeitar toda aliança ou pacto feito com o mal, consciente ou inconscientemente.
   - **Oração e Jejum**: Buscar a presença de Deus através da oração fervorosa e, se necessário, jejum para quebrar as forças espirituais negativas.
   - **Autoridade Espiritual**: Usar a autoridade em Cristo para ordenar que as maldições sejam quebradas e que as bênçãos sejam restauradas.
   - **Obediência**: Viver uma vida de obediência contínua à Palavra de Deus, mantendo-se fiel aos Seus mandamentos.

5. **Versículo Completo**:
   "Os céus e a terra tomo hoje por testemunhas contra vós, de que te tenho proposto a vida e a morte, a bênção e a maldição; escolhe, pois, a vida, para que vivas, tu e a tua descendência." - Deuteronômio 30:19

6. **O que nos conduz a uma vida regada de bênção e maldição?**
   - Nossas escolhas e atitudes em relação à obediência à palavra de Deus determinam se nossa vida será regada de bênçãos ou maldições. Seguir os mandamentos de Deus, viver em santidade, e manter-se fiel a Ele conduzem a uma vida cheia de bênçãos. Por outro lado, a desobediência, rebelião, e pecado persistente conduzem à maldição.

7. **Onde Encontramos Descanso e Libertação?**
   - Encontramos descanso e libertação em Jesus Cristo. Ele nos convida a lançar sobre Ele nossas cargas e preocupações, oferecendo-nos descanso para nossas almas. Em Cristo, somos libertos de toda maldição e condenação, e recebemos paz, perdão, e uma nova vida. 

   - **Referências Bíblicas**:
     - Mateus 11:28-30: "Vinde a mim, todos os que estais cansados e oprimidos, e eu vos aliviarei. Tomai sobre vós o meu jugo, e aprendei de mim, que sou manso e humilde de coração; e encontrareis descanso para as vossas almas. Porque o meu jugo é suave e o meu fardo é leve."
     - Gálatas 3:13: "Cristo nos resgatou da maldição da lei, fazendo-se maldição por nós; porque está escrito: Maldito todo aquele que for pendurado no madeiro."

# CAP. 11 Guerra Espiritual

1. **O que é Guerra Espiritual?**  
   Guerra espiritual é o confronto contínuo entre as forças espirituais do bem e do mal. Para os cristãos, é a batalha travada contra as forças malignas, como Satanás e seus demônios, que tentam desviar os fiéis do caminho de Deus. É um combate não físico, mas espiritual, onde a fé, oração, e a Palavra de Deus são as principais armas.

2. **Passos para Vencer o Inimigo**:
   - **Orar e Jejuar**: Buscar comunhão com Deus e fortalecimento espiritual.
   - **Estudar e Confessar a Palavra de Deus**: Usar as Escrituras para combater mentiras e tentações.
   - **Viver em Santidade**: Manter uma vida reta e separada do pecado.
   - **Revestir-se da Armadura de Deus**: Usar os elementos descritos em Efésios 6:11-18 para proteção espiritual.
   - **Resistir ao Diabo**: Firmar-se na fé e não ceder às tentações.
   - **Perseverança e Confiança em Deus**: Confiar na vitória que Cristo já conquistou na cruz.

3. **Relação entre as partes da Armadura de Deus (Efésios 6:11-18)**:
   - **1. O Capacete da Salvação**  
   ( ) Este protegerá sua mente, dando-lhe a segurança e a certeza de que é salvo e lhe permitirá discernir toda sorte de pensamentos negativos que geralmente são provenientes do adversário.
   - **2. O Cinturão da Verdade**  
   ( ) É viver de acordo com a Palavra de Deus, pois você um dia decidiu servi-Lo e afastou toda a mentira de sua vida.
   - **3. A Couraça da Justiça**  
   ( ) Protege seu peito, sua alma, seu coração e seus sentimentos. O inimigo não poderá causar-lhe dano, pois seu coração está protegido como um tesouro muito valioso nas mãos do Senhor.
   - **4. O Escudo da Fé**  
   ( ) Com ele, todos os dardos inflamados com o fogo do maligno que vêm sobre você serão apagados.
   - **5. A Espada do Espírito**  
   ( ) É a confissão da Palavra, para podermos vencê-lo. O sucesso da Palavra de Deus vem quando esta chega aos nossos lábios e a confessamos com autoridade.

4. **Importância do Sangue Derramado por Cristo no Calvário**:
   O sangue derramado por Cristo é fundamental porque é através dele que os pecados são perdoados, e o poder do inimigo é derrotado. O sacrifício de Jesus na cruz traz redenção e reconciliação com Deus, rompendo o domínio do pecado e oferecendo libertação aos crentes. Compreender a importância do sangue de Cristo fortalece a fé e capacita o cristão a resistir ao mal com autoridade.

5. **Estratégia Bíblica para Vencer o Inimigo**:
   A estratégia principal encontrada nas Escrituras para vencer o inimigo é a submissão a Deus, resistência ao diabo, e o uso da Palavra de Deus como arma espiritual. Isto inclui orar, jejuar, viver em santidade, e confiar no poder do Espírito Santo.

6. **O que a Cruz Representa na Guerra Espiritual?**:
   A cruz representa a vitória definitiva sobre o pecado, a morte, e Satanás. É o símbolo do sacrifício de Cristo, que desarmou os poderes das trevas e garantiu a redenção para a humanidade. Na guerra espiritual, a cruz é o lembrete de que o inimigo já foi derrotado e que os cristãos têm autoridade em Cristo para viver em vitória.

7. **Complete**:
   "Sujeitai-vos, pois, a Deus; resisti ao diabo, e ele fugirá de vós." - Tiago 4:7

# CAP. 12 A importância da Casa De Deus 
1. **Base do Crescimento do Cristão em Deus**:  
   A base do crescimento do cristão em Deus é uma combinação de fé, obediência à Palavra de Deus, oração, estudo bíblico, comunhão com outros crentes, e a prática dos ensinamentos de Cristo. O crescimento espiritual é alimentado pelo relacionamento íntimo com Deus e pela transformação contínua que o Espírito Santo opera na vida do cristão, levando-o a amadurecer na fé e viver de acordo com os princípios do Reino de Deus.

2. **O que é a Igreja?**  
   A igreja é a comunidade de crentes em Jesus Cristo que se reúnem para adorar a Deus, estudar Sua Palavra, compartilhar a fé, e servir uns aos outros. Ela é tanto um organismo espiritual, composto por todos aqueles que confessam Jesus como Senhor, quanto uma organização visível, com estruturas e práticas que facilitam a comunhão e o crescimento espiritual dos seus membros.

3. **Importância da Igreja**:  
   A igreja é importante porque é o meio pelo qual os crentes se reúnem para adorar a Deus, crescer em fé, receber ensino e encorajamento, e ser equipados para o serviço. A igreja também desempenha um papel crucial na evangelização, ajudando a disseminar o Evangelho e a demonstrar o amor de Cristo ao mundo. Além disso, ela oferece uma comunidade de apoio onde os cristãos podem crescer espiritualmente, aprender juntos, e ajudar uns aos outros em momentos de necessidade.

4. **Por que a Igreja é Conhecida como o “Corpo de Cristo”?**  
   A igreja é conhecida como o "Corpo de Cristo" porque, assim como um corpo físico, ela é composta por muitos membros diferentes, cada um com suas funções específicas, mas todos trabalhando juntos sob a liderança de Cristo, que é a cabeça. Esta metáfora enfatiza a unidade e a diversidade dentro da igreja, mostrando que todos os crentes são interdependentes e que Cristo é a fonte de vida e direção para todo o corpo.

5. **Complete**:
   "Pois também eu te digo que tu és Pedro, e sobre esta pedra edificarei a minha igreja, e as portas do inferno não prevalecerão contra ela." - Mateus 16:18

6. **Importância de Estar Envolvido nas Atividades da Igreja**:
   Estar envolvido nas atividades da igreja é importante porque isso permite ao cristão exercer seus dons espirituais, crescer em comunhão com outros crentes, e fortalecer sua fé. O envolvimento na igreja também contribui para a edificação do corpo de Cristo e permite que o indivíduo participe ativamente na missão de Deus no mundo, servindo aos outros e levando a mensagem do Evangelho.

7. **Explicação da Frase de John MacArthur**:  
   Quando John MacArthur afirma que "a Igreja é a expressão terrena da realidade celestial", ele está dizendo que a igreja, enquanto comunidade de crentes na terra, é uma manifestação visível do Reino de Deus. Através da igreja, a glória, a autoridade e os propósitos de Deus são demonstrados no mundo. A igreja reflete a comunhão que existe no céu, a adoração a Deus, e o cumprimento de Sua vontade, mostrando ao mundo um vislumbre da vida eterna e da santidade que caracterizam o Reino celestial.
