# **Guia Mangá de *ESTATÍSTICA***

### **Estatística** é o estudo de dados com o objetivo de compreender, analisar e tirar conclusões sobre uma população. Ao coletar e analisar dados de uma amostra representativa, a estatística ajuda a fazer previsões e inferências sobre a população como um todo.

**População**: Refere-se ao conjunto completo de indivíduos ou elementos sobre os quais queremos obter informações. A população pode ser muito grande (por exemplo, todos os habitantes de um país) ou específica (por exemplo, todos os funcionários de uma empresa).

**Amostra**: É um subconjunto da população, selecionado para representar a totalidade. A amostra é usada para fazer inferências sobre a população, com a expectativa de que os resultados sejam aplicáveis à população geral, desde que a amostra seja representativa e bem escolhida.

**Tipos de Dados em Estatística:**

1. **Dados Qualitativos/Categóricos**: São dados que descrevem qualidades ou categorias e não podem ser medidos numericamente. Eles são usados para identificar atributos ou características de uma amostra ou população.
   - **Exemplo**: Cor dos olhos (azul, verde, castanho), gênero (feminino, masculino), cidade de origem.

2. **Dados Quantitativos/Numericos**: São dados que podem ser medidos e expressos em números. Eles representam quantidades e permitem cálculos matemáticos.
   - **Exemplo**: Idade, altura, peso, salário.

### Subdivisões:
- **Qualitativos**:
   - **Nominal**: Categorias sem ordem lógica (exemplo: cores).
   - **Ordinal**: Categorias com uma ordem (exemplo: níveis de escolaridade).
- **Quantitativos**:
   - **Discreto**: Valores contáveis, sem frações (exemplo: número de filhos).
   - **Contínuo**: Pode assumir qualquer valor dentro de um intervalo (exemplo: altura, peso).

## **Como as respostas de múltipla escolhas são administrdas na prática**

Na prática, as respostas de múltipla escolha são administradas por meio de sistemas que organizam, analisam e armazenam os dados coletados de forma eficiente. Aqui estão os principais passos no processo de administração dessas respostas:

1. **Desenvolvimento do Questionário**:
   - A primeira etapa envolve a criação de perguntas de múltipla escolha que sejam claras e relevantes para o objetivo da pesquisa.
   - As opções devem cobrir as possíveis respostas sem causar confusão e, quando possível, incluir uma opção “Outro” ou “Nenhum dos anteriores”.

2. **Coleta de Respostas**:
   - A coleta pode ser feita via plataformas online (como Google Forms, SurveyMonkey, Typeform) ou em papel, dependendo do público-alvo e dos recursos disponíveis.
   - Ferramentas digitais facilitam a administração e organização dos dados automaticamente.

3. **Codificação das Respostas**:
   - Cada resposta de múltipla escolha é associada a um valor numérico ou categoria, o que facilita o processamento e a análise estatística.
   - Exemplo: “Sim” pode ser codificado como 1 e “Não” como 0.

4. **Armazenamento dos Dados**:
   - As respostas são armazenadas em um banco de dados ou em planilhas (como Excel), o que permite a organização das respostas por categoria.
   - Cada resposta de um participante é registrada como uma linha de dados, com cada pergunta representada em uma coluna.

5. **Análise das Respostas**:
   - Ferramentas estatísticas e de visualização de dados (como Excel, SPSS, Power BI) são usadas para analisar as respostas.
   - Métricas como frequência de respostas, porcentagens, médias e gráficos ajudam a interpretar os dados.

6. **Relatório e Interpretação dos Resultados**:
   - Os resultados das respostas de múltipla escolha são interpretados e apresentados em relatórios para identificar padrões, preferências e tendências.
   - Os insights gerados podem ajudar na tomada de decisões ou direcionamento de futuras pesquisas.

Esse processo permite que os dados de múltipla escolha sejam analisados de forma rápida e eficaz, fornecendo uma visão clara das preferências ou opiniões da população pesquisada.
